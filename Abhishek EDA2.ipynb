# QUESTIONS and ANSWER

1. Highest heart rate of students based on gender

```python
gender=df.groupby("Gender")["Heart_Rate"].max()
print(gender)
gender.plot(kind="pie",autopct="%1.f%%",startangle=90)
plt.axis("equal")
plt.show()
```

2. What is the distribution of sleep quality among students

```python

quality=df["Sleep_Quality"].value_counts().sort_index()
print(quality)
quality.plot(kind="bar",color="yellow")
plt.xlabel("sleep_quality")
plt.ylabel("Distribution of students")
plt.title("Distribution")
plt.show()

```

3. What percentage of students were getting good sleep quality

```python
sleep=df[df["Sleep_Quality"]=="Good"]
percentage=(len(sleep)/len(df))*100
print(percentage)
```

4. How does sleep quality impact stress levels

```python
average=df.groupby("Sleep_Quality")["Stress_Level_Self_Report"].mean()
print(average)
average.plot(kind="bar",color="red")
plt.xlabel("sleep_quality")
plt.ylabel("stresslevel")
plt.title("impact")
plt.show()
```

5. What is the distribution of blood pressure across different stress levels

```python
distribution=df.groupby("Stress_Level_Biosensor")["Blood_Pressure_Systolic"].describe()
print(distribution)
distribution.plot(kind='line',figsize=(10,6),color='skyblue',title='Distribution',
ylabel='bloodpressure',xlabel='stress level')
plt.show()

```

6. How many male and female students are there

```python
gender=df["Gender"].value_counts()
print(gender)
gender.plot(kind='line', marker='o', linestyle='-', color='b', title='gender')
plt.xlabel('female')
plt.ylabel('male')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

CONCLUSION

The dataset highlights key health metrics and behaviors among students. The highest heart rates were 99.65 for males and 97.31 for females. Sleep quality was reported as 47.3% good, 31.7% moderate, and 21% poor, with better sleep linked to lower stress levels. The average age was 20.96 years, with 519 males and 481 females. Blood pressure varied with stress, showing a systolic standard deviation of 14.60 mmHg and a minimum diastolic value of 60.0 mmHg. Among students, 190 were low risk, 672 moderate risk (218 with low physical activity), and 511 had heart rates over 70 bpm, emphasizing the impact of sleep, stress, and activity on health
